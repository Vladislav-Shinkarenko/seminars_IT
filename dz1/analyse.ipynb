```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import argparse as arg
```

```python
parser = arg.ArgumentParser()
parser.add_argument("a", type=int, help="Первый аргумент")
parser.add_argument("b", type=int, help="Второй аргумент")
args = parser.parse_args()
```

```python
def inp(a,b):
    dataset = pd.read_excel("C:\\Users\\Mi\\Desktop\\python\\2_semestr\\dz1\\outcome_{a}.{b}.xlsx")
    
    #print(dataset)

    dataset['Дата'] = [int(x.split()[0]) for x in dataset['Дата']]

    return dataset

dataset = inp(args.a, args.b)
```

```python
fig, ax = plt.subplots(constrained_layout=True)

sns.lineplot(
    data=dataset,
    x='Дата',
    y='Сумма',
    hue='Категория',
    ax=ax
)
ax.legend()

plt.title('ponos')

plt.show()

```
